his script is used to aggregate samples from different chunks of mESC data when SCRaPL is fitted with inflation. This script is also used for accessibility/expression components of mESC data.

```python
from IPython import display
import pandas as pd
import numpy as np
import scipy
import scipy.stats

from matplotlib import pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from tensorflow import keras

from tensorflow.keras import layers
import tensorflow_probability as tfp
import tensorflow.compat.v2 as tf
tf.enable_v2_behavior()

import pickle
from timeit import default_timer as timer
Folder = '/content/drive/MyDrive/'
```

```python
#For each one of the chunks store parameters of interest in a list.
cor_nuts_col = []
s_met_nuts_col = []
s_exp_nuts_col = []
m_met_nuts_col = []
m_exp_nuts_col = []
m_exp_nuts_col = []
inf_nuts_col = []

for prt in range(19):
    with open(Folder+'SCRaPL/Real/Results/nuts_cor_gastr_'+str(prt+1)+'_NT.pickle', 'rb') as handle:
        cor_nuts_col.append(pickle.load(handle))
    with open(Folder+'SCRaPL/Real/Results/nuts_s_met_gastr_'+str(prt+1)+'_NT.pickle', 'rb') as handle:
        s_met_nuts_col.append(pickle.load(handle))
    with open(Folder+'SCRaPL/Real/Results/nuts_s_exp_gastr_'+str(prt+1)+'_NT.pickle', 'rb') as handle:
        s_exp_nuts_col.append(pickle.load(handle))
    with open(Folder+'SCRaPL/Real/Results/nuts_m_met_gastr_'+str(prt+1)+'_NT.pickle', 'rb') as handle:
        m_met_nuts_col.append(pickle.load(handle))
    with open(Folder+'SCRaPL/Real/Results/nuts_m_exp_gastr_'+str(prt+1)+'_NT.pickle', 'rb') as handle:
        m_exp_nuts_col.append(pickle.load(handle))
    with open(Folder+'SCRaPL/Real/Results/nuts_inf_gastr_'+str(prt+1)+'_NT.pickle', 'rb') as handle:
        inf_nuts_col.append(pickle.load(handle))

    print(prt)
```

```python
#Concat all lists to aggregate all samples.
cor_nuts = tf.concat(cor_nuts_col,axis=-1)
s_met_nuts = tf.concat(s_met_nuts_col,axis=-1)
s_exp_nuts = tf.concat(s_exp_nuts_col,axis=-1)
m_met_nuts = tf.concat(m_met_nuts_col,axis=-1)
m_exp_nuts = tf.concat(m_exp_nuts_col,axis=-1)
inf_nuts = tf.concat(inf_nuts_col,axis=-1)
```

```python
#Save parameters of interest.
with open(Folder+'SCRaPL/Real/Results_col/nuts_cor_gastr_NT.pickle', 'wb') as handle:
    pickle.dump(cor_nuts, handle)
with open(Folder+'SCRaPL/Real/Results_col/nuts_m_met_gastr_NT.pickle', 'wb') as handle:
    pickle.dump(m_met_nuts, handle)
with open(Folder+'SCRaPL/Real/Results_col/nuts_m_exp_gastr_NT.pickle', 'wb') as handle:
    pickle.dump(m_exp_nuts, handle)
with open(Folder+'SCRaPL/Real/Results_col/nuts_s_met_gastr_NT.pickle', 'wb') as handle:
    pickle.dump(s_met_nuts, handle)
with open(Folder+'SCRaPL/Real/Results_col/nuts_s_exp_gastr_NT.pickle', 'wb') as handle:
    pickle.dump(s_exp_nuts, handle)
with open(Folder+'SCRaPL/Real/Results_col/nuts_inf_gastr_NT.pickle', 'wb') as handle:
    pickle.dump(inf_nuts, handle)
```
