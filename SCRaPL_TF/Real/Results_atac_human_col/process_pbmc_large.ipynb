```python
from IPython import display
import pandas as pd
import numpy as np
import scipy
import scipy.stats

from matplotlib import pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from tensorflow import keras

from tensorflow.keras import layers
import tensorflow_probability as tfp
import tensorflow.compat.v2 as tf
tf.enable_v2_behavior()

import pickle
from timeit import default_timer as timer
```

```python
tfd = tfp.distributions
tfb = tfp.bijectors
Folder = '/content/drive/MyDrive/'
```

```python
yy_acc_pd_1 = pd.read_csv(Folder+'SCRaPL/Real/Results_atac_human_col/Seurat_int_acc_tmp_1_1.csv',',',header=[0],index_col=[0])
yy_exp_pd_1 = pd.read_csv(Folder+'SCRaPL/Real/Results_atac_human_col/Seurat_int_exp_tmp_1_1.csv',',',header=[0],index_col=[0])

yy_acc_pd_2 = pd.read_csv(Folder+'SCRaPL/Real/Results_atac_human_col/Seurat_int_acc_tmp_1_2.csv',',',header=[0],index_col=[0])
yy_exp_pd_2 = pd.read_csv(Folder+'SCRaPL/Real/Results_atac_human_col/Seurat_int_exp_tmp_1_2.csv',',',header=[0],index_col=[0])

yy_acc_pd_3 = pd.read_csv(Folder+'SCRaPL/Real/Results_atac_human_col/Seurat_int_acc_tmp_1_3.csv',',',header=[0],index_col=[0])
yy_exp_pd_3 = pd.read_csv(Folder+'SCRaPL/Real/Results_atac_human_col/Seurat_int_exp_tmp_1_3.csv',',',header=[0],index_col=[0])
```

```python
gns_tmp = pd.read_csv(Folder+'SCRaPL/Real/Results_atac_human_col/gns_tmp_1.csv',',',header=[0])
pks_tmp = pd.read_csv(Folder+'SCRaPL/Real/Results_atac_human_col/pks_tmp_1.csv',',',header=[0])
```

```python
yy_acc_pd = pd.concat([yy_acc_pd_1,yy_acc_pd_2,yy_acc_pd_3],axis=0)
yy_exp_pd = pd.concat([yy_exp_pd_1,yy_exp_pd_2,yy_exp_pd_3],axis=0)
```

```python
print(gns_tmp)
print(pks_tmp)
```

```python
pks_unq, pks_unq_ind = np.unique(pks_tmp,return_index=True)
pks_unq = pks_unq[np.argsort(pks_unq_ind)]

gns_unq,gns_unq_ind =  np.unique(gns_tmp,return_index=True)
gns_unq = gns_unq[np.argsort(gns_unq_ind)]
```

```python
yy_acc_pd_upd = yy_acc_pd.iloc[np.argsort(pks_unq_ind),:]
yy_acc_pd_upd.index = pks_unq
yy_exp_pd_upd = yy_exp_pd.iloc[np.argsort(gns_unq_ind),:]
yy_exp_pd_upd.index = gns_unq
```

```python
yy_acc_pd_upd.to_csv(Folder+'SCRaPL/Real/Results_atac_human_col/Human_acc_seur_tmp_1.csv',sep=",")
yy_exp_pd_upd.to_csv(Folder+'SCRaPL/Real/Results_atac_human_col/Human_exp_seur_tmp_1.csv',sep=",")
```
