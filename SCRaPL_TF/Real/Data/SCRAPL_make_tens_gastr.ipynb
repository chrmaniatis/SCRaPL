This script converts methylation/expression or accessibility/expression from list fromat and to matrices in mESC data.

```python
from IPython import display
import pandas as pd
import numpy as np
import scipy
import scipy.stats
import tensorflow.compat.v2 as tf
Folder = '/content/drive/MyDrive/'
```

```python
Gastr_data=pd.read_csv(Folder+'SCRaPL/Real/Data/Gastrulation_SCRaPL_NT.csv',",", header=None)
nrm_data = pd.read_csv(Folder+'SCRaPL/Real/Data/nrm_NT.csv',",")

gene_data = pd.read_csv(Folder+'SCRaPL/Real/Data/Gastrulation_SCRaPL_gns_NT.csv',",")
```

```python
n_rows = np.max(Gastr_data.iloc[:,3])
n_cols = np.max(Gastr_data.iloc[:,4])

ind = np.asarray(np.stack([Gastr_data.iloc[:,3],Gastr_data.iloc[:,4]],axis=-1),dtype=np.int64)
print(n_rows)
print(n_cols)
```

```python
Mt = np.zeros((n_rows,n_cols))
Mt[ind[:,0]-1,ind[:,1]-1] = np.asarray(Gastr_data.iloc[:,0],dtype=np.float64)

CpG = np.zeros((n_rows,n_cols))
CpG[ind[:,0]-1,ind[:,1]-1] = np.asarray(Gastr_data.iloc[:,1],dtype=np.float64)

Rna = np.zeros((n_rows,n_cols))
Rna[ind[:,0]-1,ind[:,1]-1] = np.asarray(Gastr_data.iloc[:,2],dtype=np.float64)

Mt_pd = pd.DataFrame(Mt,index=gene_data['feature'],columns=nrm_data['cell'])
CG_pd = pd.DataFrame(CpG,index=gene_data['feature'],columns=nrm_data['cell'])
Rn_pd = pd.DataFrame(Rna,index=gene_data['feature'],columns=nrm_data['cell'])
```

```python
Mt_pd.to_csv(Folder+'SCRaPL/Real/Data/Met_NT.csv', header=True)
CG_pd.to_csv(Folder+'SCRaPL/Real/Data/CpG_NT.csv', header=True)
Rn_pd.to_csv(Folder+'SCRaPL/Real/Data/Rna_NT.csv', header=True)
```
