This script converts accessibility/expression from list format used by previous SCRaPL version and converts them into matrix format. 

```python
from IPython import display
import pandas as pd
import numpy as np
import scipy
import scipy.stats
import tensorflow.compat.v2 as tf
Folder = '/content/drive/MyDrive/'
```

```python
ATAC_data=pd.read_csv(Folder+'SCRaPL/Real/Data/dt_SCRaPL125.csv',"\t", header=0)
nrm_data = pd.read_csv(Folder+'SCRaPL/Real/Data/norm_fact_SCRaPL125.csv',"\t")
```

```python
cell_labels,ind  = np.unique(ATAC_data['cell'],return_inverse=True)
ATAC_data['cell'] = ind
```

```python
n_rows = np.max(ATAC_data.iloc[:,2])
n_cols = np.max(ATAC_data.iloc[:,3])+1

ind = np.asarray(np.stack([ATAC_data.iloc[:,2],ATAC_data.iloc[:,3]],axis=-1),dtype=np.int64)
```

```python
print(nrm_data)
```

```python
Acc = np.zeros((n_rows,n_cols))
Acc[ind[:,0]-1,ind[:,1]] = np.asarray(ATAC_data.iloc[:,0],dtype=np.float64)

Rna = np.zeros((n_rows,n_cols))
Rna[ind[:,0]-1,ind[:,1]] = np.asarray(ATAC_data.iloc[:,1],dtype=np.float64)

Acc_pd = pd.DataFrame(Acc,columns=nrm_data['cell'])
Rna_pd = pd.DataFrame(Rna,columns=nrm_data['cell'])
```

```python
Acc_pd.to_csv(Folder+'SCRaPL/Real/Data/Acc_atac.csv', header=True)
Rna_pd.to_csv(Folder+'SCRaPL/Real/Data/Rna_atac.csv', header=True)
nrm_data.to_csv(Folder+'SCRaPL/Real/Data/Nrm_atac.csv', header=True)
```
