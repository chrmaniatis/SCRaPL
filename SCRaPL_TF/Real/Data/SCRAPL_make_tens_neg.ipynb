This script is used to construct negative control data for methylation/expression or accessibility/ expression pairs in mEBC data.

```python
from IPython import display
import pandas as pd
import numpy as np
import scipy
import scipy.stats
import tensorflow.compat.v2 as tf
from matplotlib import pyplot as plt
Folder = '/content/drive/MyDrive/'
```

```python
ii = 4 #Negative control label
Gastr_data=pd.read_csv(Folder+'SCRaPL/Real/Data/Gastrulation_SCRaPL_NT_neg'+str(ii)+'.csv',",",header=None)
nrm_data = pd.read_csv(Folder+'SCRaPL/Real/Data/Gastrulation_SCRaPL_nrm_NT_neg'+str(ii)+'.csv',",")
cls_data = pd.read_csv(Folder+'SCRaPL/Real/Data/nrm_NT.csv',",")

gene_data = pd.read_csv(Folder+'SCRaPL/Real/Data/Gastrulation_SCRaPL_gns_NT_neg'+str(ii)+'.csv',",")
```

```python
n_rows = np.max(Gastr_data.iloc[:,3])
n_cols = np.max(Gastr_data.iloc[:,4])

ind = np.asarray(np.stack([Gastr_data.iloc[:,3],Gastr_data.iloc[:,4]],axis=-1),dtype=np.int64)
```

```python
Mt = np.zeros((n_rows,n_cols))
Mt[ind[:,0]-1,ind[:,1]-1] = np.asarray(Gastr_data.iloc[:,0],dtype=np.float64)

CpG = np.zeros((n_rows,n_cols))
CpG[ind[:,0]-1,ind[:,1]-1] = np.asarray(Gastr_data.iloc[:,1],dtype=np.float64)

Rna = np.zeros((n_rows,n_cols))
Rna[ind[:,0]-1,ind[:,1]-1] = np.asarray(Gastr_data.iloc[:,2],dtype=np.float64)

Norm = np.zeros((n_rows,n_cols))
Norm[ind[:,0]-1,ind[:,1]-1] = np.asarray(nrm_data.iloc[:,0],dtype=np.float64)

Mt_pd = pd.DataFrame(Mt,index=gene_data['feature'],columns=cls_data['cell'])
CG_pd = pd.DataFrame(CpG,index=gene_data['feature'],columns=cls_data['cell'])
Rn_pd = pd.DataFrame(Rna,index=gene_data['feature'],columns=cls_data['cell'])
Nrm_pd = pd.DataFrame(Norm,index=gene_data['feature'],columns=cls_data['cell'])
```

```python
Mt_pd.to_csv(Folder+'SCRaPL/Real/Data_perm/Met_NT_neg'+str(ii)+'.csv', header=True)
CG_pd.to_csv(Folder+'SCRaPL/Real/Data_perm/CpG_NT_neg'+str(ii)+'.csv', header=True)
Rn_pd.to_csv(Folder+'SCRaPL/Real/Data_perm/Rna_NT_neg'+str(ii)+'.csv', header=True)
Nrm_pd.to_csv(Folder+'SCRaPL/Real/Data_perm/nrm_NT_neg'+str(ii)+'.csv', header=True)
```
