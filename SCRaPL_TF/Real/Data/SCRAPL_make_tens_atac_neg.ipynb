This script is used to construct negative control datasets from mEBC data.

```python
from IPython import display
import pandas as pd
import numpy as np
import scipy
import scipy.stats
import tensorflow.compat.v2 as tf
from matplotlib import pyplot as plt
Folder = '/content/drive/MyDrive/'
```

```python
ATAC_data=pd.read_csv(Folder+'SCRaPL/Real/Data/dt_SCRaPL125.csv',"\t", header=0)
nrm_data = pd.read_csv(Folder+'SCRaPL/Real/Data/norm_fact_SCRaPL125.csv',"\t")
neg = 5 #Negative control dataset label
```

```python
cell_labels,ind  = np.unique(ATAC_data['cell'],return_inverse=True)
ATAC_data['cell'] = ind
```

```python
n_rows = np.max(ATAC_data.iloc[:,2])
n_cols = np.max(ATAC_data.iloc[:,3])+1

ind = np.asarray(np.stack([ATAC_data.iloc[:,2],ATAC_data.iloc[:,3]],axis=-1),dtype=np.int64)
```

```python
Acc = np.zeros((n_rows,n_cols))
Acc[ind[:,0]-1,ind[:,1]] = np.asarray(ATAC_data.iloc[:,0],dtype=np.float64)

Rna = np.zeros((n_rows,n_cols))
Rna[ind[:,0]-1,ind[:,1]] = np.asarray(ATAC_data.iloc[:,1],dtype=np.float64)

Norm_acc = np.zeros((n_rows,n_cols))
Norm_rna = np.zeros((n_rows,n_cols))
Nrm_data = nrm_data.to_numpy()
```

```python
Acc_prm = Acc.copy()
Rna_prm = Rna.copy()
Norm_acc_prm = Norm_acc.copy()
Norm_rna_prm = Norm_rna.copy()
```

```python
#Iterates through 
for ii in range(Acc_prm.shape[0]):
    map_acc = np.random.permutation(Acc_prm.shape[1])
    map_rna = np.random.permutation(Acc_prm.shape[1])
    
    Rna_prm[ii,:] = Rna_prm[ii,map_rna]
    Acc_prm[ii,:] = Acc_prm[ii,map_acc]

    Norm_acc_prm[ii,:] = Nrm_data[map_acc,1]
    Norm_rna_prm[ii,:] = Nrm_data[map_rna,2]
```

```python
Acc_prm_pd = pd.DataFrame(Acc_prm,columns=nrm_data['cell'])
Rna_prm_pd = pd.DataFrame(Rna_prm,columns=nrm_data['cell'])
Norm_acc_prm_pd = pd.DataFrame(Norm_acc_prm,columns=nrm_data['cell'])
Norm_rna_prm_pd = pd.DataFrame(Norm_rna_prm,columns=nrm_data['cell'])
```

```python
Acc_prm_pd.to_csv(Folder+'SCRaPL/Real/Data_perm/Acc_atac_neg'+str(neg)+'.csv', header=True)
Rna_prm_pd.to_csv(Folder+'SCRaPL/Real/Data_perm/Rna_atac_neg'+str(neg)+'.csv', header=True)
Norm_acc_prm_pd.to_csv(Folder+'SCRaPL/Real/Data_perm/nrm_atac_neg'+str(neg)+'.csv', header=True)
Norm_rna_prm_pd.to_csv(Folder+'SCRaPL/Real/Data_perm/nrm_rna_neg'+str(neg)+'.csv', header=True)
```
