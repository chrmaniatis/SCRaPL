```python
from IPython import display
import pandas as pd
import numpy as np
import scipy
import scipy.stats

from matplotlib import pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from tensorflow import keras
from sklearn.neighbors import KernelDensity

from tensorflow.keras import layers
import tensorflow_probability as tfp
import tensorflow.compat.v2 as tf
tf.enable_v2_behavior()

import pickle
from timeit import default_timer as timer
Folder = '/content/drive/MyDrive/SCRaPL/'
```

```python
#Load posterior correlation samples from negative control data.
ng = 5
with open(Folder+'Demo/Results/nuts_cor_300_100_'+str(ng)+'.pickle', 'rb') as handle:
    cor_nuts = pickle.load(handle)
```

```python
#Set γ to be the 90th percent qunatile of posterior correlation in absolute terms.  
smps,gns = tf.shape(cor_nuts)
col_smps = tf.reshape(cor_nuts,shape=(smps*gns))
gam = tfp.stats.quantiles(tf.math.abs(tf.math.tanh(col_smps/2)),num_quantiles=10,interpolation='linear')[-2]
```
