Efficinecy analysis for SCRaPL as a function of problem size.

```python
from IPython import display
import pandas as pd
import numpy as np
import scipy
import scipy.stats

from matplotlib import pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from tensorflow import keras

from tensorflow.keras import layers
import tensorflow_probability as tfp
import tensorflow.compat.v2 as tf
tf.enable_v2_behavior()

import pickle
from timeit import default_timer as timer
```

```python
tfd = tfp.distributions
tfb = tfp.bijectors
Folder = '/content/drive/MyDrive/'
time= np.asarray([137.67,175.67,594.42,2025.00,5947.35,16161.08],dtype=np.float32) #Time until completion
genes = np.asarray([100,500,1000,2000,3000,4000],dtype=np.int32) #Number of genes
cells = np.asarray([50,100,500,1000,2000,4000],dtype=np.int32) #Number of cells
x_data = genes*cells
labels = []
for ii in range(len(time)):
    labels.append(str( genes[ii] )+"gns./"+str(  cells[ii]  )+"cls.")
print(labels)
```

```python
#Plot recorded times
plt.plot(np.log10(x_data),np.log10(time),'-x')
plt.ylabel("log(time)(s)",fontsize=20)
plt.xticks(np.log10(x_data), labels, rotation=50,fontsize=20)
plt.savefig(Folder+'SCRaPL/Synth/Figures/Efficiency.pdf', bbox_inches='tight')
plt.show()
```
